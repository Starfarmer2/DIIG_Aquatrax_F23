```python
#small leaks pattern across days
import matplotlib.pyplot as plt
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)


days = 20
cmap = get_cmap(days)
g_data = []
# colors = ['r','g','b','orange']
for i in range(days):
    g_data.append(df_degraded_extra[df_degraded_extra['Meternumber']==meter_number][i*24:(i+1)*24])
    g = seaborn.lineplot(data=g_data[i],x="Usagedate",y="Value",color=cmap(i), label=f'Day {i+1}')
```
